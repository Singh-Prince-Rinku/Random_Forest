```python
from sklearn import datasets
```

```python
iris =datasets.load_iris()
```

```python
iris
```

```python
print(iris.target_names)
print(iris.feature_names)   
```

```python
import pandas as pd 
data = pd.DataFrame({'sepal length ':iris.data[:,0],
             'sepal width':iris.data[:,1],
             'petal length':iris.data[:,2],
             'petal width':iris.data[:,3],
             'species':iris.target})
```

```python
data.head()
```

```python
x = iris['data']
print(x)
```

```python
y = data['species']
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x,y, test_size=0.3)
```

```python
print("length of X train : ",len(X_train))
print("length of x test : ",len(X_test))
```

```python
print("length of y train : ",len(y_train))
print("length of y test : ",len(y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier= RandomForestClassifier(n_estimators=100,criterion='gini')
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.predict(X_test)
```

```python
classifier.score(X_test,y_test)
```

```python
classifier.predict([[2,3,4,5]])
```

```python
pd.Series(classifier.feature_importances_,index=iris.feature_names).sort_values(ascending=False)
```
